```python
import pandas as pd
import sqlalchemy as db

engine = db.create_engine("mssql+pyodbc://@cloudcds_ts",echo=False)
connection = engine.connect()
metadata = db.MetaData(engine,schema='tgl_hsio')
```

```python
joinID_view_t = db.Table('batched_execution_joinID_view', metadata, autoload=True, autoload_with=engine)
```

```python
q = db.select([joinID_view_t]).where(db.and_(joinID_view_t.columns.testid == '1FF73B41-282F-4D86-A1DD-6DF55D81DAE7'))
data = connection.execute(q).fetchall()
joinID_view_t = pd.DataFrame(data)
joinID_view_t.columns = data[0].keys()
joinID_view_t = joinID_view_t.rename(columns={'testid': 'TestId'}).copy()
joinID_view_t
```

check list of unique testIDs:

```python
testIDs_before = list(joinID_view_t['TestId'])
testIDs = list(joinID_view_t['TestId'].drop_duplicates())
len(testIDs_before) - len(testIDs)
```

```python
PerLaneVertical_t = db.Table('radphy_PerLaneVertical', metadata, autoload=True, autoload_with=engine)
#q = db.select([PerLaneVertical_t]).where(db.and_(PerLaneVertical_t.columns.TestId.in_(testIDs),PerLaneVertical_t.columns.JoinID=='1'))
q = db.select([PerLaneVertical_t]).where(PerLaneVertical_t.columns.TestId.in_(testIDs))
data = connection.execute(q).fetchall()
PerLaneVertical_t = pd.DataFrame(data)
PerLaneVertical_t.columns = data[0].keys()
PerLaneVertical_t
```

```python
PerLaneVertical_t.Interface.fillna(value='dummy', inplace=True)
PerLaneVertical_t.Port.fillna(value='dummy', inplace=True)
PerLaneVertical_t.Lane.fillna(value='dummy', inplace=True)
PerLaneVertical_t
```

Keeping type and BlockName information which will be joined later to the table 

```python
tmp_t = pd.concat([PerLaneVertical_t['TestId'], PerLaneVertical_t['SubTestId'], PerLaneVertical_t['JoinID'], PerLaneVertical_t['Interface'], PerLaneVertical_t['Port'], PerLaneVertical_t['Lane'], PerLaneVertical_t['Type'],PerLaneVertical_t['BlockName']],axis = 1)
tmp_t = tmp_t.drop_duplicates(subset=['TestId','SubTestId','JoinID','Interface','Port','Lane'])
tmp_t = tmp_t.drop(columns=['SubTestId','Interface','Port','Lane'])
tmp_t
```

###### Get Per Lane Veritical Table where Split By( :Key )

```python
#change Port and interface to remove conflict in converting the pivoted object into DataFrame 
#as these columns gets present when keys become columns
df_plv_split = PerLaneVertical_t.rename(columns={"Port":"Port_val","Interface":"Interface_val"}).pivot_table(index=['TestId', 'SubTestId', 'JoinID', 'Interface_val','Port_val', 'Lane'],columns="Key",values="Value",aggfunc=lambda x: ' '.join(x))
df_plv_split = df_plv_split.reset_index()
df_plv_split.rename(columns={"Port_val":"Port","Interface_val":"Interface"})
df_plv_split
```

Need to do left outer join of df_plv_split and tmp_t to gather information for type and BlockName

```python
df_plv_split.index = tmp_t.index
plv_final = pd.concat((df_plv_split,tmp_t), axis =1)
col_list = list(plv_final.columns)
col_list.insert(6, col_list.pop(col_list.index('BlockName')))
col_list.insert(6, col_list.pop(col_list.index('Type')))
#col_list
plv_final = plv_final.reindex(columns=col_list)
plv_final
```
